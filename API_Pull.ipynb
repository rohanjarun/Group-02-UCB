## US Spending by Award

```python
for i in range(1,50):
    
    url = "https://api.usaspending.gov/api/v2/search/spending_by_award"

    payload = json.dumps({
        "page": i,
        "filters": {
            "award_type_codes": ["IDV_A", "IDV_B", "IDV_B_A"],
            "recipient_scope": "domestic",
            "recipient_type_names": [
                "category_business"
            ],
            "place_of_performance_scope": "domestic",
            "award_amounts": [
                {
                    "lower_bound": 1,
                    "upper_bound": 1600000000
                }
            ],
            "time_period": [
                {
                    "start_date": "2020-01-01", 
                    "end_date": "2021-01-01"
                }
            ],
        },
        "fields": [
            "Award ID",
            "Recipient Name",
            "Start Date",
            "End Date",
            "Award Amount",
            "Awarding Agency",
            "Awarding Sub Agency",
            "Award Type",
            "Funding Agency",
            "Funding Sub Agency",
            "Loan Value"
        ],
        "sort": "Start Date",
        "order": "desc"
    })
    headers = {
        'Content-Type': 'application/json',
        'Cookie': 'cookiesession1=678A3E0DCDEFGHIJKLNOPQRSTUV08936'
    }

    response = requests.request("POST", url, headers=headers, data=payload)

    data = response.json()
    temp_df = pd.DataFrame(data['results'])
    if i == 1:
        temp_df.to_csv('Resources/Data/sample_data.csv', mode='a', index=False, header=True)
    else: temp_df.to_csv('Resources/Data/sample_data.csv', mode='a', index=False, header=False)
    print(f'Page {i} added to csv')
    
```
